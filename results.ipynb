```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure
import tableprint
import ipyplot
import glob, os
from PIL import Image


plt.rcParams["figure.figsize"] = (13,8)
```

```python
num_layers = np.arange(1,21) 
acc = [67.13, 74.76, 78.33, 80.14, 81.23, 81.66, 80.90, 80.59, 76.54, 74.31, 48.16, 46.65, 10, 10, 10, 10, 10, 10, 10, 10]

num_iters = [1,10, 20] 
acc_1 = [64, 67.13, 60.73 ]
acc_2 = [70.28, 73.42, 74.2]
acc_3 = [71.70 ,74.42, 73.83]

plt.figure()
plt.plot(num_layers, acc, 'b', label='feed-forward')
plt.plot(num_iters, acc_1, 'r', label='DT_recall, alpha=0')
plt.plot(num_iters, acc_2, 'g', label='DT_recall, alpha=0.5')
plt.plot(num_iters, acc_3, 'y', label='DT_recall, alpha=1')
plt.title('Feed forward Vanilla Conv Net on Cifar10')
plt.xlabel('Number of Layers (recurrence)')
plt.ylabel('Accuracy')
plt.grid()
plt.legend()
plt.show()
```
